```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

#Function to Load data from a CSV file 
def load_data(file_path): 
    return pd.read_csv(file_path)

#Function to explore the dato (print first few rows, data info, and summary statistics)

def explore_data(df):

    print("First few cours of the dataset:")

    print(df.head())

    print("\nData information:")

    print(df.info())

    print("\nSummary statistics:")

    print(df.describe())

# Function to plot the breakdown of distance from home by job role and attrition



def plot_distance_by_job_role (df):

    distance_by_job_role =df.groupby(['JobRole', 'Attrition'])['DistanceFromHome'].mean().reset_index() 
    distance_pivot=distance_by_job_role.pivot(index='JobRole', columns='Attrition', values='DistanceFromHome')
    plt.figure(figsize=(10, 6))
    sns.barplot(data=distance_pivot, x=distance_pivot.index, y='Yes', color='red', label='Attrition - Yes')
    sns.barplot(data=distance_pivot, x=distance_pivot.index, y='No', color='blue', bottom=distance_pivot ['Yes'], label='Attrition - No') 
    plt.xticks(rotation=90)

    plt.xlabel('Job Role')

    plt.ylabel('Average Distance from Home')

    plt.title('Breakdown of Distance from Home by Job Role and Attrition')

    plt.legend()

    plt.show()
           
 #Function to plot the comparison of average monthly income by education and attritionJupyterLab

def plot_income_by_education (df):
        
            income_by_education =df.groupby(['Education', 'Attrition'])['MonthlyIncome'].mean().reset_index() 
            income_pivot= income_by_education.pivot(index='Education', columns='Attrition', values='MonthlyIncome')
            plt.figure(figsize=(10, 6))
            sns.barplot(data=income_pivot, x=income_pivot.index, y='Yes', color='green', label='Attrition - Yes')
            sns.barplot(data=income_pivot, x=income_pivot.index, y='No', color='brown', bottom=income_pivot ['Yes'], label='Attrition - No') 
            plt.xticks(rotation=90)
            plt.xlabel('Education') 

            plt.ylabel('Average Monthly Income')

            plt.title('Comparison of Average Monthly Income by Education and Attrition')

            plt.legend()

            plt.show()
        
if __name__ == "__main__":

#file path for the CSV data

 file_path = 'WA_Fn-UseC_-HR-Employee-Attrition.xls'

#Step 1: Load the doto

data=load_data(file_path)

#Step 2: Explore the dato

explore_data(data)


        
```

```python
# Step 3: Plot the breakdown of distance from home by job role and attrition
plot_distance_by_job_role(data)
```

```python
# Step 4: Plot the comparison of average monthly income by education and attrition 
plot_income_by_education(data)
```
